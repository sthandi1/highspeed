```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, rfft, rfftfreq
```

```python
t = np.linspace(0, 2*np.pi, 10000)
```

```python
f = 1
x = np.sin(2*np.pi*f*t)
```

```python
fig, ax = plt.subplots()
ax.plot(t,x)
ax.set_xlabel('t (seconds)')
ax.set_ylabel('x (metres)')
ax.axvline(x=0, color='grey')
ax.axhline(y=0, color='grey')
```

```python
xfft = fft(x)
xfftfreqs = fftfreq(len(x), 2*np.pi/len(x))
calculated_freqs = []
fs = len(x)/(2*np.pi)
for i in range(len(xfft)):
    bin_freq = i*fs/(len(xfft))
    calculated_freqs.append(bin_freq)

freqs_linspace = np.linspace(0, fs, len(xfft), endpoint=False)
```

```python
fig1, ax1 = plt.subplots()
ax1.stem(xfftfreqs, xfft)
ax1.set_xlim(-2,2)
```

```python
fig2, ax2 = plt.subplots()
ax2.stem(calculated_freqs, xfft)
ax2.set_xlim(-2,2)
```
